# 02 Exploratory Analysis

This notebook explores Maryland county access risk patterns, risk drivers, and state-average comparisons.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

PROJECT_ROOT = Path.cwd().resolve().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
data_path = PROJECT_ROOT / 'data' / 'processed' / 'dashboard_county_risk.csv'
df = pd.read_csv(data_path, dtype={'county_fips': str})
df.head()
```

```python
df[['access_risk_score', 'poverty_pct', 'uninsured_pct', 'provider_access_index', 'chronic_burden_index']].describe().round(2)
```

```python
top_risk = df.sort_values('access_risk_score', ascending=False).head(10)
top_risk[['county_name', 'region', 'access_risk_score', 'access_risk_tier', 'provider_gap_index', 'socioeconomic_need_index', 'chronic_burden_index']]
```

```python
ax = top_risk.sort_values('access_risk_score').plot.barh(
    x='county_name',
    y='access_risk_score',
    legend=False,
    figsize=(8, 5),
    color='#c1121f'
)
ax.set_title('Top Maryland Counties by Access Risk Score')
ax.set_xlabel('Risk score')
ax.set_ylabel('')
plt.tight_layout()
```

```python
state_average = df.select_dtypes(include='number').mean(numeric_only=True)
comparison = df.assign(
    risk_vs_state_avg=df['access_risk_score'] - state_average['access_risk_score'],
    poverty_vs_state_avg=df['poverty_pct'] - state_average['poverty_pct'],
    uninsured_vs_state_avg=df['uninsured_pct'] - state_average['uninsured_pct'],
)
comparison[['county_name', 'risk_vs_state_avg', 'poverty_vs_state_avg', 'uninsured_vs_state_avg']].sort_values('risk_vs_state_avg', ascending=False).head(10).round(1)
```
